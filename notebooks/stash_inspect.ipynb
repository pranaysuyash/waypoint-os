```python
# Inspect Git stash object for recovery
from pathlib import Path
import zlib

git_dir = Path('/Users/pranay/Projects/travel_agency_agent/.git')
stash_hash = '44057f6ab706ce818cff81ddbaae1906d51d2cb9'
obj_path = git_dir / 'objects' / stash_hash[:2] / stash_hash[2:]
print('object path:', obj_path)
with obj_path.open('rb') as f:
    data = zlib.decompress(f.read())
print('raw header:', data[:50])
print('full header until null:', data.split(b'\x00', 1)[0])
# Dump first 200 bytes after header
parts = data.split(b'\x00', 1)
if len(parts) > 1:
    print('content sample:', parts[1][:200])
else:
    print('no content after header')
```
